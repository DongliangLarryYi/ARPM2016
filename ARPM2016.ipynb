```python
import numpy as np 
import pandas as pd
import pandas_datareader.data as web
import datetime as dt
```

```python
SP500 = pd.read_csv('./SPY_All_Holdings.csv')
SP500['Weight'] = SP500['Weight']/100
SP500
```

```python
prices = pd.DataFrame()
start = dt.datetime(2016,1,1)
tickers = SP500['Ticker']
for ticker in tickers:
    prices[ticker] = web.DataReader(ticker, 'yahoo', start)['Adj Close']
returns = prices.pct_change()
returns = returns.cumsum()
returns
```

```python
cols=['Consumer Discretionary', 'Consumer Staples', 'Energy', 'Financials', 'Health Care', 'Industrials', 'Information Technology', 'Materials', 'Telecommunication Services', 'Utilities']
factors = pd.DataFrame(data=0, columns=cols, index=returns.index)
for column in returns:
    for row in SP500.iterrows():
        if column == row[1][1]:
            if row[1][3] == 'Consumer Discretionary':
                factors['Consumer Discretionary'] = factors['Consumer Discretionary'] + returns[column]
            elif row[1][3] == 'Consumer Staples':
                factors['Consumer Staples'] = factors['Consumer Staples'] + returns[column]
            elif row[1][3] == 'Energy':
                factors['Energy'] = factors['Energy'] + returns[column]
            elif row[1][3] == 'Financials':
                factors['Financials'] = factors['Financials'] + returns[column]
            elif row[1][3] == 'Health Care':
                factors['Health Care'] = factors['Health Care'] + returns[column]
            elif row[1][3] == 'Industrials':
                factors['Industrials'] = factors['Industrials'] + returns[column]
            elif row[1][3] == 'Information Technology':
                factors['Information Technology'] = factors['Information Technology'] + returns[column]
            elif row[1][3] == 'Materials':
                factors['Materials'] = factors['Materials'] + returns[column]
            elif row[1][3] == 'Telecommunication Services':
                factors['Telecommunication Services'] = factors['Telecommunication Services'] + returns[column]
            elif row[1][3] == 'Utilities':
                factors['Utilities'] = factors['Utilities'] + returns[column]
factors
```

```python
sectors = SP500.groupby('Sector')
sectors = sectors.sum()
sectors
```
